# Importing required libraries

```python
# Importing the essential libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Loading the dataset

```python
#Loading the dataset
data = pd.read_csv("C://nhl_elo_latest.csv")
```

# Lineplot with multiple lines

```python
# Creating line plot with multiple lines
x = ['Nashville Predators vs San Jose Sharks','San Jose Sharks vs Nashville Predators','New York Rangers vs Tampa Bay Lightning','Los Angeles Kings vs Vegas Golden Knights','Montreal Canadiens vs Toronto Maple Leafs']
y1 = np.array([1506.26777290077, 1446.46438083354, 1550.56881835361, 1501.12906760357, 1439.3127407971])
y3 = np.array([1450.24086317589, 1510.04425524312, 1571.15890583938, 1528.45437452053, 1554.43502260661])
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y1, label= 'home team win probablity')
ax.plot(x, y3, label='away team win probablity')
ax.set_xticks(x)

# Rotate the axis of labels by 45 degree
ax.set_xticklabels(['Nashville Predators vs San Jose Sharks','San Jose Sharks vs Nashville Predators','New York Rangers vs Tampa Bay Lightning','Los Angeles Kings vs Vegas Golden Knights','Montreal Canadiens vs Toronto Maple Leafs'], rotation=45)

# Adding labels, title and legend in the plot
ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
ax.set_title('Probablity of win the match between the home team and away team')
ax.legend()

# Displaying the plot
plt.show()
```

# Pie Chart of home team score

```python
labels = ['Nashville Predators ','San Jose Sharks ','New York Rangers ','Los Angeles Kings ','Montreal Canadiens ']
values = np.array([4, 2,3, 3, 4])
```

```python
# Developing the pie plot
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%')

# Adding a title in the pie plot
ax.set_title('Home Team Score Distribution')

# Showing the plot
plt.show()
```

# Scatter plot

```python
# extracting the values of x and y  from the dataset
x = data['home_team_postgame_rating']
y = data['away_team_postgame_rating']

# creating the scatter plot
plt.scatter(x, y, color='Magenta')

# adding plot labels and plot title
plt.xlabel('home team rating(post-game)')
plt.ylabel('away team rating(post-game)')
plt.title('Scatter Plot between the post match ratings of the home team and away team')

# Displaying the scatter plot
plt.show()
```
